```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os

A = imread('13.webp')
X = np.mean(A,-1)

U, S, VT = np.linalg.svd(X,full_matrices=False)

column_corr = X.T@X
row_corr = X.dot(X.T)

plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.title('Korelacja - wiersze')
plt.imshow(row_corr, cmap='viridis')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.title('Korelacja - kolumny')
plt.imshow(column_corr, cmap='viridis')
plt.colorbar()

plt.show()
```
